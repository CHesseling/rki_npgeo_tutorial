# RKI Daten abrufen 1

Jetzt nehmen wir uns die RKI-Daten im Detail vor. Das Gute daran: Sie liegen im CSV-Format vor, das wir leicht mit Pandas öffnen können. Zunächst muss man das Pandas Modul in Python laden:


```
import pandas as pd
```

Das ist die URL zu dem RKI-Datensatz. Die URL ist jeden Tag zum Glück gleich. Damit die Spalten "IdLandkreis" und "IdBundesland" als Text erkannt werden, können wir vor dem Laden der CSV Tabelle bestimmen, welchen Datentyp die Spalten haben sollen. Außerdem können wir gleich bestimmen, dass die Spalten "RefDatum", "Meldedatum" und "Datenstand" als Typ Datumsangaben enthalten und als solche "geparst" - also verarbeitet - werden sollen.



```
dtypes = {'IdLandkreis': str, 'IdBundesland': str}
df = pd.read_csv('https://opendata.arcgis.com/datasets/dd4580c810204019a7b8eb3e0b329dd6_0.csv', parse_dates=['Meldedatum', 'Refdatum', 'Datenstand'], dtype=dtypes)
```

Jetzt können wir den DataFrame aufrufen:



```
df
```

```
df.tail(10)
```

- `int64` bedeutet Integer

```
df.info()
```

Und mit dem Befehl .describe() lassen sich auf die Schnelle ein paar statistische Auswertungen darstellen:

# Dataframe filtern<br>
Dataframes lassen sich anhand einzelner Spalten filtern. Dazu kann man Zahlen vergleichen (zum Beispiel mit `<`, `>` oder `==`) oder Textspalten filtern (mit `==`):


```
df['Bundesland'] == 'Nordrhein-Westfalen'
```

Dadurch wird ein Wahrheitswert zurück gegeben. Will man nur die Zeilen, bei denen der Wahrheitswert `True`lautet, muss man noch `df[(  )]` um den Filterausdruck herumsetzen. Und dann die Wert in einen neuen Dataframe speichern:


```
df_nrw = df[(df['Bundesland'] == 'Nordrhein-Westfalen')]
df_nrw
```

## Pivot Tabellen

Mit Pivot Tabellen kann man die Spalten und Zeilen neu "überkreuzen", deshalb werden sie auch Kreuztabellen genannt. Es ist eines der mächtigsten Tools, um große Datenmengen zu verstehen.
<br>
Um in Pandas eine Pivot Tabelle zu bauen, braucht man folgende Zutaten
 1. Quell-DF 
 2. Index = Was soll in den Zeilen / in der "horizontalen" Achse angezeigt werden
 3. columns = sollen die Daten noch in Spalten aufgefächert werden?
 4. values = Was soll gezählt werden? Welche Spalte im DF ist es?
 5. aggunc = Wie soll es gezählt werden (sum, mean, ect.)


```
pd.pivot_table(df_nrw, index='Altersgruppe', columns="Geschlecht", values='AnzahlFall', aggfunc='sum')
```

# Zellen ansteuern mit `iloc`<br>

Wenn man aus einem Dataframe ein bestimmten Wert extrahieren möchte, kann man dazu den "Integer Locator" nutzen - also den Locator nach Zahlen (gemeint sind die Werte für Zeile und Spalte).<br>
Damit kann man unterschiedliche Dinge machen. Angenommen, ich habe einen Dataframe nach Meldedatum und den Bundesländern in den Spalten:

```
df_bl_nach_datum = pd.pivot_table(df, index='Meldedatum', columns="Bundesland", values='AnzahlFall', aggfunc='sum')
```

```
df_bl_nach_datum
```

Mit `.iloc[0]` lässt sich die erste Zeile anzeigen. (Wie immer in Python 0 = 1. Wert)<br>
Fachwissen: In diesem Fall ist das eine Pandas Serie (und kein Dataframe)

```
df_bl_nach_datum.iloc[0]
```

Um das ganze als Dataframe anzuzeigen, kann man die erste Zeile bis exklusive der 2. Zeile anzeigen:

```
df_bl_nach_datum.iloc[0:1]
```

```
Genau so lässt sich die letzte Zeile anzeigen:
```

```
# Als Serie
df_bl_nach_datum.iloc[-1]
```

```
# Als Dataframe
df_bl_nach_datum.iloc[-1:]
```

Man kann sich so zum Beispiel die letzten 7 Tage des Dataframes anzeigen lassen:

```
df_bl_nach_datum.iloc[-7:]
```

Bislang bekommen wir hier die vollen Zeilen. Wenn man nur einen oder ausgewählte Spalten anzeigen lassen möchte, geht das ebenfalls mit `.iloc` - die Spaltenwerte folgen nach dem Komma in den eckigen Klammern:

```
df_bl_nach_datum.iloc[-1:,4:6]
```

...oder, um nur einen Wert zu extrahieren:

```
df_bl_nach_datum.iloc[-1,4:5].item()
```

Neben dem Integer-Locator gibt es auch noch Namens-Locator `.loc()`

```
df_bl_nach_datum.loc[:,'Nordrhein-Westfalen']
```

`loc` sowohl für Zeilen als auch Spalten:

```
df_bl_nach_datum.loc['2020-11-17','Nordrhein-Westfalen']
```

Da sich in der Automation aber immer das neueste Datum ändert, kann man auch `loc` und `iloc` kombinieren:

```
df_bl_nach_datum.loc[:,'Nordrhein-Westfalen'].iloc[-1]
```

# Rolling

Es gibt praktische Pandas-Funktionen, um zum Beispiel die Summe der letzten 7 Tage zu berechnen. Das geht mit `rolling`:
- Wie viele Tage sollen zur Berechnung herangezogen werden?
- min_periods= Wie viele Tage sollen minimal beobachtet werden (z.B. wenn die Zahl der Tage nicht glatt durch 7 teilbar ist)?
- Axis = Soll das ganze per Zeile (1) oder per Spalte (0) berechnet werden?
- Schließlich muss noch bestimmt werden, ob man die Summe oder etwas anderes berechnen will (`.sum()`)


```
df_bl_nach_datum.rolling(7, min_periods=7, axis=0).sum()
```
